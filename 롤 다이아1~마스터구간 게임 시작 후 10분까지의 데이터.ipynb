```python
#머신러닝
from sklearn import model_selection
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
x=data.iloc[:,[16,17]].values
y=data['blueWins'].values
```

```python
x,y
```

```python
train_x,test_x,train_y,test_y=model_selection.train_test_split(x,y,test_size=0.3)
```

```python
fit=LogisticRegression().fit(x,y)
```

```python
y_predict=fit.predict(train_x)
train_score=metrics.accuracy_score(train_y,y_predict)
print("훈련 데이터 정확도 : {0:.3f}".format(train_score))
```

```python
y_test=fit.predict(test_x)
test_score=metrics.accuracy_score(test_y,y_test)
print("검증 데이터 정확도 : {0:.3f}".format(test_score))
```

블루팀의 골드차이, 경험치차이로 승리여부를 결정짓기 위해 변수 x에서 골드차이와 경험치차이을 뽑고, y에서 승리여부를 뽑아 7:3으로 train_test데이터를 나눈 뒤 로지스틱회귀분석으로 예측하였다.
훈련데이터로 승패여부를 예측한 결과 정확도가 72.6%였고 검증데이터 정확도가 72.9%로 약간의 과소적합이 있었지만 별 차이는 없었다.

```python
confmat=pd.DataFrame(confusion_matrix(test_y,y_test),index=['True[0]','True[1]'],
                    columns=['Predict[0]','Predict[1]'])
```
